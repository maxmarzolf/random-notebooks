```python
import pandas as pd
```

```python
with open('data/transcript.txt', "r") as word_list:
    words = word_list.read().split(' ')
words = pd.DataFrame(words, columns=['text'])
words
```

```python
words['text'] = words['text'].str.replace(r'[^\w\s]+', '')
```

```python
names = ['Daniel', 'Alarcón', 'Juan', 'Manuel', 'Robles']
words = words[~words.text.isin(names)]
words = words[~words.text.str.contains('\n\n\n\n')]
words = words['text'].value_counts()
words = words[:100]
```

```python
words.plot.bar(figsize=(20,10));
```
